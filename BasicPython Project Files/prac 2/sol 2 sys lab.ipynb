```python
#printing pythagorean triplets
count=1
import math
for a in range(1,500,1):
    for b in range(a,500,1):
        c=math.sqrt((a*a)+(b*b))
        if c in range(1,500,1):
            print("PYTHAGOREAN TRIPLET NO. {} IS :".format(count))
            print("{}-{}-{}".format(a,b,int(c)))
            count=count+1
            
```

```python
#point inside or not cheaker
def area_calculator(x1, y1, x2, y2, x3, y3): 
    return abs((x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2.0) 
   
def whether_point_lies_or_not(x1, y1, x2, y2, x3, y3, x, y):
    #for triangle ABC 
    A = area_calculator (x1, y1, x2, y2, x3, y3) 
    #for triangle PBC  
    A1 = area_calculator (x, y, x2, y2, x3, y3)       
    #for triangle PAC  
    A2 = area_calculator (x1, y1, x, y, x3, y3) 
    #for triangle PAB  
    A3 = area_calculator (x1, y1, x2, y2, x, y) 
    if(A == A1 + A2 + A3): 
        return 1
    else: 
        return 0

k=whether_point_lies_or_not(10, 10, 12, 5, 7, 12, 70, 70)    
if (k==1): 
    print('POINT IS INSIDE OF ABC') 
else: 
    print('POINT IS OUTSIDE OF ABC')  
  
```

```python
#program to find sum of series
x=input("Enter the value of variable.: ")
x=float(x)
n=input("Enter the degree of n.: ")
n=int(n)
accum=0
for i in range(0,n+1,1):
    if (i==0):
        accum=accum+1
        print(accum)
    elif (i==1):
        accum=accum+0
        print(accum)
    else:
        accum=accum+((x**i)/(i))
        print(accum)
    n=n+1
        
print("Sum of series is {}".format(accum))       
```
