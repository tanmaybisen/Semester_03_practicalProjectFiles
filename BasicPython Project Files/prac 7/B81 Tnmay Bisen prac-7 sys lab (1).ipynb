```python
# Q4) View basic statistical details like percentile, mean, std deviation, mode, variance, skewness and kutosis of Automobile data                                       

import pandas 

dataset = pandas.read_csv("Automobile_data.csv",na_values=["?",0])
dataset = {'%tile':dataset.quantile(),'MEAN':dataset.mean(),'Standard deviation':dataset.std(),'MODE':dataset.mode(),'VAR':dataset.var(),'SKEW':dataset.skew(),'KURT':dataset.kurt(),'MEDIAN':dataset.median(),'Count-Result':dataset.count()}       
df = pandas.DataFrame(dataset)
print(df)
```

```python
# Q5) Find the most expensive car company name

import pandas
data = pd.read_csv("Automobile_data.csv",na_values=["?",0])
val=data[['make','price']][data.price==data['price'].max()]
print(val)
```

```python
# Q6) Print all Toyata cars available

import pandas

data1 = pd.read_csv("Automobile_data.csv")
data1=data1[data1['make']=="toyota"]
print(data1)
```

```python
# Q7)  Find each companyâ€™s highest price car
data = pandas.read_csv("Automobile_data.csv",na_values=["?",0])
company = data.groupby('make')
h_price = company['make','price'].max()
print(h_price)
```

```python
# Q8)  Find the average mileage of each car making company

import pandas

data = pandas.read_csv("Automobile_data.csv",na_values=["?",0])
company = data.groupby('make')
avg_m = company['make','city-mpg'].mean()
print(avg_m)
```

```python
# Q9)  Sort all cars by price column

import pandas 
data = pandas.read_csv("Automobile_data.csv",na_values=["?",0])
data = data.sort_values(by=['price'], ascending=True)
print(data)
```
