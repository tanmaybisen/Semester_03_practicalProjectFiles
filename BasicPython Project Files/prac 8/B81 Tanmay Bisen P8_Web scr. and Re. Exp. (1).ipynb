```python
# Web page 
# https://bongirwarvk.webnode.com/wat/
    
#     Find out all links
#     find text
#     find how many times Test/assignment/lecture word is appeared on to the page


#solution 1 finding all links


import requests
from bs4 import BeautifulSoup
url = 'https://bongirwarvk.webnode.com/cgngui/'
r = requests.get(url)
htmlContent =  r.content

soup = BeautifulSoup(htmlContent, 'html.parser')
soup
soup.prettify

links=set()
x=soup.find_all('a')
#print(x)
for link in x:
    links.add(link.get('href')) #append(link.get('href'))
links
```

```python
import requests
from bs4 import BeautifulSoup as bs

ph = requests.get("https://bongirwarvk.webnode.com/wat/")
soup = bs(ph.content,'html.parser')
text = soup.get_text()
print("printing TEXT \n")
print(text)
```

```python
# solution 3 finding how many times 
# Test/assignment/lecture word
# is appeared on to the page
import re

count=0
for string in soup.strings:
    p=string.split()
    for h in p:
        x=re.findall("^(A|a)ssignment$",h)
        count=count+len(x)

count2=0
for string in soup.strings:
    p=string.split()
    for h in p:
        x=re.findall("^(t|T)est$",h)
        count2=count2+len(x)
        
count3=0
for string in soup.strings:
    p=string.split()
    for h in p:
        x=re.findall("^(L|l)ecture$",h)
        count3=count3+len(x)
        
print("Number of times assignment word appeared")
print(count)
print("Number of times Test word appeared")
print(count2)
print("Number of times lecture word appeared")
print(count3)

```

```python
# Regular Expression Regex:
# 1.Check Valid Email
while(a!='stop'):
    a=input("enter email id : ")
    if a!='stop':
        x = re.findall(".+@.+\..+", a)
        if x:
            print("valid")
        else:
            print("invalid")
```

```python
# 2.Check Valid Mobile no.(Start with 7 or 8 or 9, 10 digit no.)
a='1';
while(a!='0'):
    a=input("enter number : ")
    if a!='0':
        x = re.findall("[7-9][0-9]{9}", a)
        if x:
            print("valid")
        else:
            print("invalid")
```

```python
# 3.Complex number { 2+3j,  2.9+3.5j, 0+4j, 4+0j}
import re
a='112'
while(a!='stop'):
    a=input("enter complex no. : ")
    if a!='stop':
        x = re.findall("^[0-9]*\.*[0-9]*\+[0-9]*\.*[0-9]*j$", a)
        if x:
            print("valid")
        else:
            print("invalid")
```

```python
# 4.Decimal number {3, 3.8, 0.77}
import re
a='112'
while(a!='stop'):
    a=input("enter decimal any number : ")
    if a!='stop':
        x = re.findall("^[0-9]*\.[0-9]*$", a)
        if x:
            print("valid")
        else:
            print("invalid")
```
